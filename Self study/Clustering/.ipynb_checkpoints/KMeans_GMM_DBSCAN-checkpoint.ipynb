```python
# Constants
DATA_PATH = "burger_king_items.txt"
```

```python
# Import Packages
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

#from plotnine import *
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.cluster import DBSCAN

from sklearn.metrics import silhouette_score
```

___
## KMeans:

```python
df = pd.read_csv(DATA_PATH, sep='\t')
df.head()
```

```python
features= ['Fat(g)', 'Sodium(mg)']
X = df[features]
scaler = StandardScaler()

X[['Fat(g)', 'Sodium(mg)']] = scaler.fit_transform(X)

# note: X = scaler.fit_transform(X) ==> gives X as nparray

#(ggplot(X, aes("Fat(g)", "Sodium(mg)")) + geom_point())
sns.scatterplot(x='Fat(g)', y='Sodium(mg)', data=df, color='k')
plt.show()
```

```python
km = KMeans(n_clusters = 3)
km.fit(X)

membership = km.predict(X)

X["cluster"] = membership

(ggplot(X, aes("Fat(g)", "Sodium(mg)", color = "membership")) + geom_point())
```

```python
silhouette_score(X[features], membership)
```

### Multiple features:

```python
features = ['Fat(g)', 'Sodium(mg)', 'Calories', 'Chol(mg)',
           'Sugar(g)', 'Protein(g)']
X = df[features]

scaler = StandardScaler()
X[features] = scaler.fit_transform(X)

km = KMeans(n_clusters = 5)
km.fit(X)

membership = km.predict(X)
X['cluster'] = membership

silhouette_score(X[features], membership)
```

___
# EM and Gaussian Mixture Models:

```python
features = ['Calories', 'Sodium(mg)']
X = df[features]

scaler = StandardScaler()
X[features] = scaler.fit_transform(X)

GMM = GaussianMixture(n_components=3)
GMM.fit(X)

cluster_prob = GMM.predict_proba(X)
clusters = GMM.predict(X)
X['clusters'] = clusters

silhouette_score(X[features], clusters)
```

```python
cluster_prob[:5, :]
```

```python
ggplot(X, aes(x='Calories', y='Sodium(mg)', color='clusters')) + geom_point()
```

```python
sns.scatterplot(x='Calories', y='Sodium(mg)', data = X, hue='clusters')
plt.show()
```

### Multiple Features:

```python
features = ['Fat(g)', 'Sodium(mg)', 'Calories', 'Chol(mg)',
           'Sugar(g)', 'Protein(g)']
X = df[features]

scaler = StandardScaler()
X[features] = scaler.fit_transform(X)

GMM = GaussianMixture(n_components=3)
GMM.fit(X)

cluster_proba = GMM.predict_proba(X)
clusters = GMM.predict(X)
X['clusters'] = clusters

silhouette_score(X[features], clusters)
```

```python
cluster_proba[:5, :]
```

```python
(ggplot(X, aes(x = "Sugar(g)", y = "Fat(g)", color = "clusters")) + geom_point())
```

```python
(ggplot(X, aes(x = "Sugar(g)", y = "Protein(g)", color = "clusters")) + geom_point())
```

___
# DBSCAN:

```python
features = ['Calories', 'Sodium(mg)']
X = df[features]

scaler = StandardScaler()
X[features] = scaler.fit_transform(X)

dbscan = DBSCAN(eps=0.5, min_samples=10)
dbscan.fit(X)

clusters = dbscan.labels_
X['clusters'] = clusters

silhouette_score(X[features], clusters)
```

```python
(ggplot(X, aes(x='Calories', y='Sodium(mg)', color='clusters')) + geom_point())
```

```python
sns.scatterplot(x='Calories', y='Sodium(mg)', data=X, hue='clusters')
plt.show()
```

### Multiple Features

```python
features = ['Fat(g)', 'Sodium(mg)', 'Calories', 'Chol(mg)',
           'Sugar(g)', 'Protein(g)']
X = df[features]

scaler = StandardScaler()
X[features] = scaler.fit_transform(X)

dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(X)

clusters = dbscan.labels_
X['clusters'] = clusters

silhouette_score(X[features], clusters)
```

```python
(ggplot(X, aes(x = "Sugar(g)", y = "Fat(g)", color = "clusters")) + geom_point())
```

## Notes:
1) KMeans is the simplest and the fastest one but isn't the best
2) GMM is more accurate and use soft assignment
* **Both of them are Centroid/Prametric clustering techniques, So they are affected by outliers.**
* **They can work with high dimensional data and deal with overlapping/touching clusters.**

3) DBSCAN is best for dense data and for detecting outliers so not assigned into any cluster.
* **It's a Density-based/Non-Parametric algorithm.**
* **but it sufers from the curse of dimensionality problem; we can see that from the value of silhouette_score.**
* **Not great with overlapping/touching clusters.**
* **Suboptimal when clusters have different densities.**

**Choosing Min_samples:**
* Domain Knowledge + Distance metrics
* More rows = larger min_pts
* More noise = larger min_pts
* More features = larger min_pts

**Choosing Epsilon:**
* Domain Knowledge
* Elbow method (K-dist)
