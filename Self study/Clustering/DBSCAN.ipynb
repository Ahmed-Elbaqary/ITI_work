```python
# Constants
DATA_PATH = "iris_pca.csv"
```

```python
# Packages
import pandas as pd
import numpy as pn

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import DBSCAN
```

___
## Processing

```python
df = pd.read_csv(DATA_PATH)
```

```python
dbscan = DBSCAN(eps=1, min_samples=5)
dbscan.fit(df)
```

```python
cluster = dbscan.labels_
df['cluster'] = cluster
```

```python
sns.scatterplot(x='pc1', y='pc2', data=df, hue='cluster')
plt.show()
```

```python
df = pd.read_csv(DATA_PATH)
```

```python
dbscan = DBSCAN(eps=1.5, min_samples=1)
dbscan.fit(df)
```
