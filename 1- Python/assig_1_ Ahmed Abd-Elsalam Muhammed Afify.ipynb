```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs=32000
ts=1/fs
l=1024
t=np.arange(l)*ts
fc=0.25*(fs/2)
x=np.sin(2*np.pi*t*fc)
fc=0.5*(fs/2)
x+=0.5*np.sin(2*np.pi*t*fc)
plt.plot(t,x)
```

```python
def myDFT(x):
    #https://en.wikipedia.org/wiki/Discrete_Fourier_transform
    N = len(x)
    n = np.arange(N)    
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    X = np.dot(e, x)
    return X

def DFT_shift(x):
    #shift rotate the victor for better visialiation
    return x[0:len(x)]
```

```python
y=myDFT(x)
f=np.linspace(-1,1,len(y))
plt.plot(f,DFT_shift(np.abs(y)))
```
