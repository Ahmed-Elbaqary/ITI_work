```python
class moving_obj():
    def __init__(self):
        self.v=0
        self.vdirction=0 
        #0>+y, 1>+x , 2>-y,  3>-x
        self.x=0
        self.y=0
    
    def update(self, command):
        #location update
        self.x=self.x+(self.vdirction==1)*self.v-(self.vdirction==3)*self.v
        self.y=self.y+(self.vdirction==0)*self.v-(self.vdirction==2)*self.v
        '''if self.vdirction==0:
            self.y=self.y+self.v
        elif self.vdirction==1:
            self.x=self.x+self.v
        elif self.vdirction==2:
            self.y=self.y-self.v
        elif self.vdirction==3:
            self.x=self.x-self.v'''
        
        #velocity update
        if command=='w':
            self.v+=1
        elif command=='s':
            self.v-=1
        elif command=='a':
            self.vdirction=(self.vdirction-1)%4
        elif command=='d':
            self.vdirction=(self.vdirction+1)%4
        
                    
    
    def display(self):
        dir_dict={0:'+y', 1:'+x' , 2:'-y',  3:'-x'}
        print('v =',(self.v,dir_dict[self.vdirction]))
        print('pos=',(self.x, self.y))
        
        
```

```python
class player(moving_obj):
    def __init__(self,name):
        self.name=name
        super().__init__()
    #target=(x,y)
    def iswon(self,target):
        if ((self.x,self.y)==target):
            return True
        else:
            return False
        
        
    
```

```python
class enemy(moving_obj):
    def __init__(self,name,):
        self.name=name
        super().__init__()
    #target=(x,y)
    def iswon(self,target):
        if ((self.x,self.y)==target):
            return True
        else:
            return False
```

```python
target=(int(input('target_x=')), int(input('target_y=')))
n=int(input('num of player = '))
p=[player(input('Enter your Name:')) for i in range(n)]
counter=0
while True:
    #general display
    counter+=1
    print('=======',target,'=================== w,s,d,a,end')
    #read and update
    flage=False
    for i in range(n):
        command=input(p[i].name)
        if command=='end':
            flage=True
        p[i].update(command)
        print('----')
    if flage:
        break
    #display players status
    for i in range(n):
        print(p[i].name)
        p[i].display()
    # show winners
    flag=False
    for i in range(n):
        if p[i].iswon(target):
            print('***********')
            print(p[i].name)
            print('Congrates******')
            print('nom of steps = ',counter)
            flag=True
    
    if flag:
        break
 
    if counter==100:
        break
        
    
```
