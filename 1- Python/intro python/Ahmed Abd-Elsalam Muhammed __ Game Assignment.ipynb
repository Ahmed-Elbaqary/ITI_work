```python
class moving_obj():
    def __init__(self):
        self.v=0
        self.vdirction=0 
        self.x=0
        self.y=0
    
    def update(self, command):
        #location update
        self.x=self.x+(self.vdirction==1)*self.v-(self.vdirction==3)*self.v
        self.y=self.y+(self.vdirction==0)*self.v-(self.vdirction==2)*self.v
        
        #velocity update
        if command=='w':
            self.v+=1
        elif command=='s':
            self.v-=1
        elif command=='a':
            self.vdirction=(self.vdirction-1)%4
        elif command=='d':
            self.vdirction=(self.vdirction+1)%4
        
    def display(self):
        dir_dict={0:'+y', 1:'+x' , 2:'-y',  3:'-x'}
        print('v =',(self.v,dir_dict[self.vdirction]))
        print('pos=',(self.x, self.y))
        
        
```

```python
class player(moving_obj):
    def __init__(self,name):
        self.name=name
        super().__init__()
    
    #target=(x,y)
    def iswon(self,target):
        if ((self.x,self.y)==target):
            return True
        else:
            return False
        
        
    
```

```python
class enemy(moving_obj):
    def __init__(self,name,):
        self.name=name
        super().__init__()
    #target=(x,y)
    def iswon(self,target):
        if ((self.x,self.y)==target):
            return True
        else:
            return False
```

```python
target=(int(input('target_x=')), int(input('target_y=')))
n=int(input('No. of player = '))
m=int(input('No. of enemies = '))
p=[player(input(f'Enter Player{i+1} Name:')) for i in range(n)]
e=[enemy(input(f'Enter Enemy{i+1} Name:')) for i in range(m)]
counter=0

while True:

    #_________________________ Players & Enemies movements ___________________________ 
    counter+=1
    print('===================',target,'=================== Movements (w,s,d,a) or (end)')
    # _______________________________________ Players movements _______________________________________
    flage=False
    out_players = []
    for i in range(len(p)):
        command=input(f"Player {p[i].name}'s Move: ")
        p[i].update(command)
        if command=='end':
            out_players.append(i)
            
    # _________ Delete players that entered end ___________________-
    for i in out_players:
        del p[i]

    # _______________________________________ Check if All Players Quited _______________________________________
    if not p:
        flag = True
    if flage:
        break
    
    # ___________________________________ Enenmies Movements ___________________________________
    out_enemies = []
    for i in range(len(e)):
        command=input(f"Enemy {e[i].name}'s Move: ")
        e[i].update(command)
        if command=='end':
            out_enemies.append(i)
    for i in out_enemies:
        del e[i]

    # _______________________ If Distance is 1 then a player looses _______________________
    lost_players = []
    for i in range(len(p)):
        for j in range(len(e)):
            if p[i].x in [e[j].x-1, e[j].x+1] and p[i].y in [e[j].y-1, e[j].y+1]:
                lost_players.append(i)
    for i in lost_players:
        del p[i]
   # _______________________________________ Check if all players looses _______________________________________ 
    if not p:
        flag = True
    if flage:
        break

    # _______________________________________ Display results after each movement _______________________________________
    print('Players: ')
    for i in range(len(p)):
        print(f"{i} Enemy {p[i].name} results")
        p[i].display()
    
    print("Enimies: ")
    for i in range(len(e)):
        print(f"{i} Enemy {e[i].name} results")
        e[i].display()
    
    
    # __________________________ show winners ______________________________________________
    flag=False
    for i in range(len(p)):
        if p[i].iswon(target):
            print('________________________________________________________________________')
            print(f"player {p[i].name} has won")
            print('****************** Congratulations ******************')
            print(f'you have reached the target in {counter} steps')
            flag=True
    
    for i in range(len(e)):
        if e[i].iswon(target):
            print('________________________________________________________________________')
            print(f"Enemy {e[i].name} has won")
            print('****************** Good Luck Next time ******************')
            flag=True
    # ____________________ if a winner, End the game ______________________
    if flag:
        break
    # ________________________________________ Check for No. of steps _______________________
    if counter==100:
        break
```
