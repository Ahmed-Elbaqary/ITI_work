```python
import numpy as np
import matplotlib.pyplot as plt 
import scipy 
```

```python
n = 800
A = 1.995653
B = 1.27689
C = 8
```

```python
r=np.linspace(0, 1, num=n)
theta=np.linspace(-2,20*np.pi,n)
```

```python
R,THETA=np.meshgrid(r, theta, indexing='xy')
```

```python
#define the number of petals we want per cycle. Roses have 3 and a bit.
petalNum=2.6
x = 1 - (1/2)*((5/4)*(1 - np.remainder(petalNum*THETA, 2*np.pi)/np.pi)**2 - 1/4)**2
phi = (np.pi/2)*np.exp(-THETA/(C*np.pi))
y = np.sin(phi)*A*(R**2)*(B*R - 1)**2
R2 = x*(R*np.sin(phi) + y*np.cos(phi))
X=R2*np.sin(THETA)
Y=R2*np.cos(THETA)
Z=x*(R*np.cos(phi)-y*np.sin(phi))
```

```python
#define a red map for our rose colouring
red_map=np.linspace(1,0.25,num=10)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, color='r')

plt.show()
```
