## Step_3:

```python
# Specify data:
x = df.drop(['Outcome'], axis=1)
y = df['Outcome']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
```

```python
# first splitting the data:
x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
clf = LogisticRegression(random_state=0)
clf.fit(x_train, y_train)
```

```python
print(f'training set error: {clf.score(x_train, y_train)}')
print(f'test set error: {clf.score(x_valid, y_valid)}')
```

```python
inter = clf.intercept_
```

```python
coef = clf.coef_
```

```python
y_hat = clf.predict(x_valid)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(f'Accuracy Score is {accuracy_score(y_valid,y_hat)}')
```

```python
from sklearn.metrics import confusion_matrix,classification_report
conf_matrix = confusion_matrix(y_valid,y_hat)
print(f'confusion_matrix: \n {conf_matrix}')
```

```python
sns.heatmap(confusion_matrix(y_valid,y_hat), annot=True)
plt.show()
```

```python
precision = 99/(99+20)

from sklearn.metrics import plot_precision_recall_curve
disp = plot_precision_recall_curve(clf, x_train, y_train)
disp.ax_.set_title('2-class Precision-Recall curve: '
                  'AP={0:0.2f}'.format(precision))
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
ns_probs = [0 for _ in range(len(y_train))]
# predict probabilities
lr_probs = clf.predict_proba(x_train)
# keep probabilities for the positive outcome only
lr_probs = lr_probs[:, 1]
# calculate scores
ns_auc = roc_auc_score(y_train, ns_probs)
lr_auc = roc_auc_score(y_train, lr_probs)
# summarize scores
print('No Skill: ROC AUC=%.3f' % (ns_auc))
print('Logistic: ROC AUC=%.3f' % (lr_auc))
# calculate roc curves
ns_fpr, ns_tpr, _ = roc_curve(y_train, ns_probs)
lr_fpr, lr_tpr, _ = roc_curve(y_train, lr_probs)
# plot the roc curve for the model
plt.plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')
plt.plot(lr_fpr, lr_tpr, marker='.', label='Logistic')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

## Implementation of Logistic regression:

```python
def Logistic_GA(x, y, alpha=0.1, epoches=1000):
    # initialize the parameters:    
    y.to_numpy().reshape(-1, 1)
    m = x.shape[0]
    Xs = np.concatenate((np.ones((m,1)), x), axis=1)
    W = np.ones((Xs.shape[1],1))
    ind_y = (y==1).to_numpy().reshape(-1,1)            # indicator function in the log-likelihood function
    
    #loss = []
    
    # Implementing Iterations:
    for i in range(epoches):
        score = Xs @ W
        sigmoid_prob = 1 / (1 + np.exp(-score)).reshape(-1,1)
    
        y_sig = ind_y - sigmoid_prob
        #log_like = -((1-ind_y)*(score)) - (np.log(1/sigmoid_prob))
        grd = (Xs.T @ y_sig).reshape(-1,1)
        #error = -np.mean(y@(np.log(sigmoid_prob))-(1-y)@np.log(1-sigmoid_prob))
        
        #loss.append(error)
        W = W + alpha*grd

    prediction = (grd>= 0.5)*1
    
    return W, prediction#, loss
```

```python
W, prediction = Logistic_GA(x_train, y_train, 0.001, 100)
```
