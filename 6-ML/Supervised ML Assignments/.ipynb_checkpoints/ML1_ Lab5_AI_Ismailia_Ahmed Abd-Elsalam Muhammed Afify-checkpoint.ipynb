```python
scores=[mean_absolute_error, mean_absolute_percentage_error, mean_squared_error, median_absolute_error, r2_score]
```

## Linear regression

```python
x = pd.DataFrame(df['Temperature (C)'])
y = pd.DataFrame(df['Apparent Temperature (C)'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
for score in scores:
    print(f'Score on training set: {score(y_train,model.predict(x_train))}')
    print(f'score on test set: {score(y_test,model.predict(x_test))}')
    print()
```

```python
# Showing Regression Coefficient
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

```python
sm.OLS(y_train, x_train).fit().summary()
```

## Then we try Multiple Linear regression

```python
x = pd.DataFrame(df[['Temperature (C)','Humidity','Visibility (km)','Pressure (millibars)']])
y = pd.DataFrame(df['Apparent Temperature (C)'])
```

## Then we try Polynomial Linear regression

we see that if we increase the order of **Humidity** and **Visibility (km)** may represent the **Apparent Temperature (C)** better

At First we try to find the best degree for **Humidity** with **Apparent Temperature (C)**

```python
x = pd.DataFrame(df['Humidity'])
y = pd.DataFrame(df['Apparent Temperature (C)'])
for deg in range(2,6):
    poly_regs= PolynomialFeatures(degree= deg)  
    x_poly= poly_regs.fit_transform(x) 
    sm.OLS(y_train, x_train).fit().summary()
```

```python
x = pd.DataFrame(df['Visibility (km)'])
y = pd.DataFrame(df['Apparent Temperature (C)'])
for deg in range(2,5):
    poly_regs= PolynomialFeatures(degree= deg)  
    x_poly= poly_regs.fit_transform(x)
    print(sm.OLS(y_train, x_train).fit().summary())
```

so we found that the best polynomial degree for Humidity is 10
