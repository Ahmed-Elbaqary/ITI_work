```python
#importing all the libraries which are neede
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import seaborn as sns
from itertools import product
%matplotlib inline
```

```python
#Read input dataset
col_names = ['sepal length','sepal width','petal length','petal width', 'class']
df = pd.read_csv('iris.data', names=col_names)
```

```python
# # Check for outliers
# Q1 = np.percentile(df[['sepal length', 'sepal width', 'petal length', 'petal width']], [25, 25, 25, 25], axis=0, interpolation='midpoint')
# Q3 = np.percentile(df[['sepal length', 'sepal width', 'petal length', 'petal width']], [75, 75, 75, 75], axis=0, interpolation='midpoint')
# IQR = Q3 - Q1
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
IQR
#filtered = df.query('(@Q1 - 1.5 * @IQR) <= "sepal width" <= (@Q3 + 1.5 * @IQR)')
```

```python
print("Old Shape: ", df.shape) 
  
# Upper bound
upper = np.where(df >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(df <= (Q1-1.5*IQR))
  
print(upper)
```

```python

df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)
  
print("New Shape: ", df.shape)
```

```python
#check for missing data
df.isnull().sum()
```

```python
#drop duplicate data instances if any
print(f'before: {df.shape}')
df.drop_duplicates(inplace=True)
print(f'before: {df.shape}')
```

```python
# Split the data into training and testing
X=df.drop('class',axis=1)
y=df['class']
```

```python
#Create Training and Test Sets and Apply Scaling
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

___
## Using Linear Kernel

```python
# Creating the model
clf1 = SVC(kernel='linear')
clf1.fit(X_train, y_train)
print('Accuracy of SVM classifier on training set: {:.2f}'
     .format(clf1.score(X_train, y_train)))
print('Accuracy of SVM classifier on test set: {:.2f}'
     .format(clf1.score(X_test, y_test)))
```

___
## Using RBF Kernel

```python
# Creating the model
clf2 = SVC(kernel='rbf')
clf2.fit(X_train, y_train)
print('Accuracy of SVM classifier on training set: {:.2f}'
     .format(clf2.score(X_train, y_train)))
print('Accuracy of SVM classifier on test set: {:.2f}'
     .format(clf2.score(X_test, y_test)))
```

___

```python
# Creating the model
clf3 = SVC(kernel='poly')
clf3.fit(X_train, y_train)
print('Accuracy of SVM classifier on training set: {:.2f}'
     .format(clf3.score(X_train, y_train)))
print('Accuracy of SVM classifier on test set: {:.2f}'
     .format(clf3.score(X_test, y_test)))
```

```python
# Split the data into training and testing
X=df.iloc[:,:2]
y=df['class']
```

```python
#Create Training and Test Sets and Apply Scaling
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

___
## Plotting decision boundaries:

```python
# Plotting decision regions

x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

f, axarr = plt.subplots(2, 2, sharex='col', sharey='row', figsize=(10, 8))

for idx, clf, tt in zip(product([0, 1], [0, 1]),
                        [clf1, clf2, clf3, clf4],
                        ['Linear kernel', 'RBF Kernel',
                         'polt Kernel', 'sigmoid Kernel']):

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    axarr[idx[0], idx[1]].contourf(xx, yy, Z, alpha=0.4)
    axarr[idx[0], idx[1]].scatter(X_train[:, 0], X_train[:, 1], c=y_train,
                                  s=20, edgecolor='k')
    axarr[idx[0], idx[1]].set_title(tt)

plt.show()
```
