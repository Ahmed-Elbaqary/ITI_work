## Data Preparing:

**In this step we do some of operations to get our data in a good quality to train our model.**
Some of these operatins are:
#### 1) Cleaning data:
**Real world data is messy so we need to clean it up, Some of popular cleaning cases:**
* **Inconsistent values:** (Just removed as they are not valid or simply replace one of them with the value of the other).
* **Duplicated records:** (Merge duplicates and drop them).
* **missing values:** (Can be filled with appropriate value, or simply dropped).
* **Invalid data:** (Just removed).
* **Outliers:** (if not in the same trend of the whole data and may cause misleading understanding for the model, just remove).

#### 2) Transforming data:
**Getting data in shape to be usefully used in the model and lead to relevant predictions**
* **Scalling, Normalization:** (Changing range of values to be within a specified values e.g.(0, 1).
* **Standerization:** (transforming data to be of standard distribution mean=0.0, std=1.0)
* **Feature selection:** (include selecting relevant features and if needed transforming some features by taking log(), srt, ...etc .
* **Dimensionality Reduction (PCA, LDA, ...):** When data has large features which are related to each other and can be reduces to some lower number of features.

### Data transformation:

**IMputing categorical data using OnehotEncoder**

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# Creating OneHotEncoder object:
OH_Encoder = OneHotEncoder()
df_ohe = df.copy()
```

```python
# Encoding the maried_single column:
maried_single = pd.DataFrame(OH_Encoder.fit_transform(df_ohe[['Married/Single']]).toarray(), columns=OH_Encoder.categories_[0])
maried_single.head()
```

```python
# Encoding the Have_House column:
Have_House = pd.DataFrame(OH_Encoder.fit_transform(df_ohe[['House_Ownership']]).toarray(), columns=OH_Encoder.categories_[0])
Have_House.head()
```

```python
# Encoding the Have_Car column:
Have_Car = pd.DataFrame(OH_Encoder.fit_transform(df_ohe[['Car_Ownership']]).toarray(), columns=OH_Encoder.categories_[0])
Have_Car.head()
```

```python
# Now merging the new columns:
df = df.merge(maried_single,  left_index=True, right_index=True)
df = df.merge(Have_House,  left_index=True, right_index=True)
df = df.merge(Have_Car,  left_index=True, right_index=True)
```

```python
df.head()
```

**Now dropping unused categorical columns**

```python
df.drop(['Married/Single', 'House_Ownership', 'Car_Ownership', 'Profession', 'CITY', 'STATE'], axis=1, inplace=True)
df.head()
```

```python
df.info()
```

**The next step is to specify our training and target data**

```python
x = df.drop('Risk_Flag', axis=1)
y = df['Risk_Flag']
```

```python
print(x.shape)
print(y.shape)
x.head()
```

**First Split data into train and test**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=42)
```

```python
print(x_train.shape)
x_train.head()
```

```python
y_train.value_counts()
```

```python
sns.catplot(x=y_train, data=y_train, kind='count')
plt.show()
```

**Now we will perform SMOTE oversampling to get over imbalanced data**

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE()
x_train_sm, y_train_sm = sm.fit_resample(x_train, y_train)
```

```python
print(y_train_sm.value_counts())
```

```python
print(x.shape)
print(y.shape)
```

**Now scalling our data**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(x_train_sm)
x_train_sm = scaler.transform(x_train_sm)
x_test = scaler.transform(x_test)
```

```python
print(x_train_sm[0,:])
```

## Model Selection:

**As we are classifying, so we can use any classification model based on the data so we will try both Logistic Regression, KNN DecesionTree and SVC with our implementations and compare them with the sklearn classifiers.**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
lo_reg = LogisticRegression()
KNN = KNeighborsClassifier(n_neighbors=5)
svc = SVC(kernel='linear')
DT = DecisionTreeClassifier(max_depth=10)
```

### Now Using cross_val_score function:

```python
from sklearn.model_selection import cross_val_score
```

***Firstly with Logistic Regression***

```python
# With impalanced data
cross_val_score(lo_reg, x_train, y_train, cv=5)
```

```python
# With SMOTE balanced data
cross_val_score(lo_reg, x_train_sm, y_train_sm, cv=5)
```

***Second model we use KNN***

```python
# With impalanced data
cross_val_score(KNN, x_train, y_train, cv=3)
```

```python
# With SMOTE balanced data
cross_val_score(KNN, x_train_sm, y_train_sm, cv=3)
```

***Last with Decision tree***

```python
# With impalanced data
cross_val_score(DT, x_train, y_train, cv=3)
```

```python
# With SMOTE balanced data
cross_val_score(DT, x_train_sm, y_train_sm, cv=3)
```
