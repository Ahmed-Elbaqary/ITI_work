```python
# --------------- Q5 -----------------
m = np.array([[1,1,0,2],[0,3,0,3],[1,0,4,4]])
lst = []
arr_sum = 0
for i in range(m.shape[1]):
    for j in range(m.shape[0]):
        if m[j,i] == 0:
            lst.append(m[j,i])
            arr_sum += m[j,i]
            break
        else:
            lst.append(m[j,i])
            arr_sum += m[j,i]

print(lst)
print(arr_sum)
```

```python
# --------------- Q6 -----------------
n = np.array([[1,1,0,2],[0,3,0,3],[1,0,4,4]])
lst = []
num = 3
for i in range(n.shape[1]):
    for j in range(m.shape[0]):
        if n[j,i] == num:
            continue
        else:
            lst.append(n[j,i])

print(lst)
```
